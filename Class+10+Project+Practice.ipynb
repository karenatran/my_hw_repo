```python
%matplotlib inline
import pandas as pd
import sklearn.linear_model as lm
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../data/flight_delays.csv')
df = df[df.DEP_DEL15.notnull()]
```

```python
df.info()
```

```python
df.head()
```

```python
df_new = pd.get_dummies(df, columns=['CARRIER','DEST_AIRPORT_ID'], drop_first=True)
```

```python
df_new.head()
```

```python
features = [i for i in df_new.columns if i.startswith('DEST_AIRPORT_ID')]
features += ['CRS_DEP_TIME']
features
```

```python
df_new[features]
```

```python
model = lm.LogisticRegression()
model.fit(df_new[features], df_new['DEP_DEL15'])
```

```python
from sklearn import metrics
```

```python
auc = metrics.roc_auc_score(df_new.DEP_DEL15, model.predict_proba(df_new[features]).T[0])
auc
```

```python
preds = pd.Series(model.predict_proba(df_new[features]).T[0])
df_new['Predictions'] = preds
```
